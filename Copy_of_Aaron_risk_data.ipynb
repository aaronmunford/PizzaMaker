```python
def identify_high_risk_areas(data, risk_columns, tract_column='TRACT'):
   """
   Identifies areas classified as 'Relatively High' or 'Very High' risk and includes tract data for comparison.


   Parameters:
   - data (pd.DataFrame): The dataset to analyze.
   - risk_columns (list): List of columns to check for risk levels.
   - tract_column (str): Name of the tract column to ensure its inclusion.


   Returns:
   - pd.DataFrame: Filtered dataframe with only high risk areas, including tract information.
   """
   high_risk_levels = ['Relatively High', 'Very High']
   high_risk_filter = data[risk_columns].apply(lambda x: x.isin(high_risk_levels)).any(axis=1)
   high_risk_areas = data[high_risk_filter]

   # Make sure the tract column is considered if further manipulations are done
   if tract_column and tract_column not in high_risk_areas.columns:
       raise ValueError(f"The specified tract column '{tract_column}' does not exist in the dataframe.")

   return high_risk_areas


# Assuming your risk datasets are already loaded and column names are checked
risk_columns = ['CFLD_EALR', 'HRCN_EALR', 'CFLD_RISKR', 'HRCN_RISKR']  # Use actual risk columns names


# Apply the function to each dataset, specifying the tract column
FL_high_risk = identify_high_risk_areas(FL_short_cleaned, risk_columns, 'TRACT')
#TX_high_risk = identify_high_risk_areas(TX_risk, risk_columns, 'TRACT')
#GA_high_risk = identify_high_risk_areas(GA_risk, risk_columns, 'TRACT')
print(FL_high_risk)
```

```python
def rank_risk_areas(df, risk_columns, weights=None):
    """
    Rank areas based on risk factors.

    :param df: DataFrame containing the risk data.
    :param risk_columns: List of columns in the df that are considered for risk assessment.
    :param weights: List of weights corresponding to the risk columns. If None, equal weighting is assumed.
    :return: DataFrame with additional 'risk_score' and 'risk_rank' columns.
    """
    # Check if weights are provided, otherwise use equal weights
    if weights is None:
        weights = [1] * len(risk_columns)

    # Standardize the risk columns
    df_standardized = df.copy()
    for col in risk_columns:
        df_standardized[col] = (df[col] - df[col].mean()) / df[col].std()

    # Calculate the weighted risk score
    df_standardized['risk_score'] = (df_standardized[risk_columns] * weights).sum(axis=1)

    # Rank the areas based on risk score
    df_standardized['risk_rank'] = df_standardized['risk_score'].rank(ascending=False)

    return df_standardized

# Example usage
risk_columns = ['RISK_SCORE', 'EAL_VALT', 'S1702_C02_001E']  # Update these columns based on your data
weights = [0.5, 0.3, 0.2]  # Example weights, adjust based on importance

FL_ranked = rank_risk_areas(FL_short_cleaned, risk_columns, weights)
#TX_ranked = rank_risk_areas(TX_risk, risk_columns, weights)
#GA_ranked = rank_risk_areas(GA_risk, risk_columns, weights)
```
